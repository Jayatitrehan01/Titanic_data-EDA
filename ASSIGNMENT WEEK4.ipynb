HANDLING MISSING VALUES

```python
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace=True)
```

```python
titanic_data.drop(columns='Cabin',inplace=True)
```

```python
titanic_data.fillna({'Age': titanic_data['Age'].median()}, inplace=True)
```

```python
print(titanic_data.isnull().sum())
```

DATA DISTRIBUTION

NUMERICAL DATA : AGE,FARE

```python
sns.histplot(titanic_data['Age'],kde=True).set(title='AGE DISTRIBUTION')
plt.show()

sns.histplot(titanic_data['Fare'],kde=True).set(title='FARE DISTRIBUTION')
plt.show()
```

CATEGORICAL DATA : SEX, PCLASS, EMBARKED

```python
sns.countplot(x=titanic_data['Sex'], data=titanic_data)
plt.title("Passenger Count by Sex")
plt.show()

sns.countplot(x='Embarked', data=titanic_data)
plt.title("Embarkation Ports")
plt.show()

sns.countplot(x='Pclass', data=titanic_data)
plt.title("Passenger Classes")
plt.show()
```

IN CONCLUSION TO THE ABOVE 3 HISTOGRAMS/COUNTPLOTS
* THERE ARE MORE MALE PASSENGERS THEN FEMALE
* MOST PEOPLE BOARDED FROM 'S' AND FEWER FROM 'Q'
* MAJORITY OF PASSENGER WERE IN 3RD CLASS AS CONPARED TO 1ST AND 2ND

DETECTING OUTLIERS USING BOXPLOT

```python
sns.boxplot(x=titanic_data['Age'])
plt.title('OUTLIERS IN AGE')
plt.show()


sns.boxplot(x=titanic_data['Fare'])
plt.title("OUTLIERS IN FARE")
plt.show()
```

RELATIONSHIP WITH VARIABLES

```python
sns.countplot(x=titanic_data['Survived'], hue='Sex', data=titanic_data)
plt.title("Survival Count by Gender")
plt.show()
```

IN CONCLUSION TO THE AOVE HISTORGRAM
* FEMALE HAD MORE SURVIVAL CHANCES AS COMPARED TO FEMALE

```python
sns.countplot(x=titanic_data['Pclass'], hue='Sex', data=titanic_data)
plt.title("Survival Count by PASSENGER CLASS")
plt.show()
```

IN CONCLUSION TO THIS HISTOGRAM
* IN ALL THE 3 CLASS, FEMALE SURVIVAL CHANCES ARE MORE THAN MALE
* IN 3RD CLASS, SURVIVAL CHANCES OF FEMALE IS VERY HIGH AS COMPARED TO MALE AND FEMALE OF ALL THE 3 CLASSES

```python
sns.boxplot(x=titanic_data['Survived'],y=titanic_data['Age'],data=titanic_data)
plt.title("Survival Count by Age ")
plt.show()
```

IN CONCLUSION TO THIS BOXPLOT
* MEDIAN AGE OF SURVIVED AND NOT SURVIVED IS BETWEEN 20-40
* MOST OUTLIERS ABOVE THE AGE OF 55 HAD NOT SURVIVED
* YOUNGER HAD SLIGHTLY BETTER SURVIVAL CHANCES

```python
plt.figure(figsize=(10,6))
sns.heatmap(titanic_data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

IN CONCLUSION TO THE CORRELATION HEATMAP
* HIGHER THE FARE, HIGHER THE SURVIVAL CHANCES
* SURVIVAL CHANCES DEPENDS ON THE PCLASS. HIGHER THE PCLASS, HIGHER THE SURVIVAL CHANCE AND VICE-VERSA
* SIBSP & PARCH HAS +VE RELATION, IT MEANS FAMILY MEMBERS OFTEN TRAVEL TOGETHER
